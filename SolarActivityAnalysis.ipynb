**Import needed modules**

```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.io import output_file, show
import os, sys, subprocess
import numpy as np
import scipy.special
import bokeh.io
import bokeh.plotting
```

**Set Environment**

**Import CSV file from GitHub**

```python
filename = 'https://raw.githubusercontent.com/lupus-in-fabula/SympatheticFlares/main/SN_ms_tot_V2.0.xls'
colnames = ['YEAR', 'MONTH', 'DATE FRAC', 'SUNSPOTNUMBER', 'STDDEVIATION', 'NUMBEROFOBSERVATIONS', 'DEFINITE']

monthlySunspotNumbers = pd.read_csv(filename, sep=';', names=colnames, header=None)
```

```python
# check the data

monthlySunspotNumbers.tail()
```

**Visualize the dataset to confirm**

```python
p = bokeh.plotting.figure(
    width=400,
    height=300,
    x_axis_label="Date Fraction",
    y_axis_label="Sun Spot Numbers (Smoothed)",
)

p.line(monthlySunspotNumbers['DATE FRAC'], monthlySunspotNumbers['SUNSPOTNUMBER'], legend_label='Sun Spot Numbers', line_width=2)
```

```python
bokeh.io.show(p)
```
